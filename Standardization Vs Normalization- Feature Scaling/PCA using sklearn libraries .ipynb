```python
data = pd.DataFrame(dataset.data,columns=dataset.feature_names)
data.head()
```

```python
data.describe()
```

```python
data.corr()
```

```python
X = data
y = dataset.target
```

```python
from sklearn.preprocessing import StandardScaler

std = StandardScaler()
X_std = std.fit_transform(X)
```

```python
X_std[:5]
```

```python
from sklearn.model_selection import train_test_split

x_train_std, x_test_std, y_train_std, y_test_std = train_test_split(X_std, y, random_state=42, test_size=0.75)
print(x_train_std.shape, y_train_std.shape)
print(x_test_std.shape, y_test_std.shape)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
lr_std = LogisticRegression(max_iter=5000)
lr_std.fit(x_train_std, y_train_std)
print(lr_std.score(x_test_std, y_test_std))
```

```python
from sklearn.decomposition import PCA

# pca = PCA(n_components=10)
pca = PCA(0.99)
X_pca = pca.fit_transform(X)
print(X_pca.shape)
```

```python
X_pca[:5]
```

```python
pca.explained_variance_ratio_
```

```python
print(pca.n_components*100,"%")
```

```python
print(pca.n_components_)
```

```python
x_train_pca, x_test_pca, y_train_pca, y_test_pca = train_test_split(X_pca, y, random_state=42, test_size=0.75)
print(x_train_pca.shape, y_train_pca.shape)
print(x_test_pca.shape, y_test_pca.shape)
```

```python
lr_pca = LogisticRegression(max_iter=5000)
lr_pca.fit(x_train_pca, y_train_pca)
print(lr_pca.score(x_test_pca, y_test_pca))
```
